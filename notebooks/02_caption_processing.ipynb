```python
import pandas as pd, re
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load captions
captions = pd.read_csv("/content/data/captions.txt")  # adjust path

# Clean & tokenize
def clean_text(text):
    text = text.lower()
    text = re.sub('[^a-z ]', '', text)
    return 'startseq ' + text + ' endseq'

captions['caption'] = captions['caption'].apply(clean_text)

tokenizer = Tokenizer(num_words=10000, oov_token="<unk>")
tokenizer.fit_on_texts(captions['caption'])
max_len = max(len(c.split()) for c in captions['caption'])

import pickle
pickle.dump(tokenizer, open("tokenizer.pkl","wb"))
```
